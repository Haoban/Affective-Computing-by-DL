### Task 4.3 Define the Network Architecture in Pytorch

The piece of code below defines the network architecture by Pytorch. This code is adopted from https://github.com/prlz77/vgg-face.pytorch/blob/master/models/vgg_face.py with sligh modification.
```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torchfile
from skimage import io

# to install the torch file,
# please first download (or git clone https://github.com/bshillingford/python-torchfile.git), 
# then install inside the repository by 'python setup.py install --user'
```

```python
# Adopted from https://github.com/prlz77/vgg-face.pytorch/blob/master/models/vgg_face.py with slight modification
# Adopted from https://github.com/prlz77/vgg-face.pytorch/blob/master/models/vgg_face.py with slight modification
# Adopted from https://github.com/prlz77/vgg-face.pytorch/blob/master/models/vgg_face.py with slight modification
# Adopted from https://github.com/prlz77/vgg-face.pytorch/blob/master/models/vgg_face.py with slight modification
# Adopted from https://github.com/prlz77/vgg-face.pytorch/blob/master/models/vgg_face.py with slight modification
# Adopted from https://github.com/prlz77/vgg-face.pytorch/blob/master/models/vgg_face.py with slight modification

class VGG_16(nn.Module):

    def __init__(self, num_classes):
        """
        Constructor
        """
        super().__init__()
        self.block_size = [2, 2, 3, 3, 3]

        ## Define Block of VGG16
        
        # Block 1
        self.conv_1_1 = nn.Conv2d(3, 64, 3, stride=1, padding=1)
        self.conv_1_2 = nn.Conv2d(64, 64, 3, stride=1, padding=1)
        
        
        # Block 2
        self.conv_2_1 = nn.Conv2d(64, 128, 3, stride=1, padding=1)
        self.conv_2_2 = nn.Conv2d(128, 128, 3, stride=1, padding=1)
        
        
        # Block 3
        self.conv_3_1 = nn.Conv2d(128, 256, 3, stride=1, padding=1)
        self.conv_3_2 = nn.Conv2d(256, 256, 3, stride=1, padding=1)
        self.conv_3_3 = nn.Conv2d(256, 256, 3, stride=1, padding=1)
        
        
        # Block 4
        self.conv_4_1 = nn.Conv2d(256, 512, 3, stride=1, padding=1)
        self.conv_4_2 = nn.Conv2d(512, 512, 3, stride=1, padding=1)
        self.conv_4_3 = nn.Conv2d(512, 512, 3, stride=1, padding=1)
        
        
        # Block 5
        self.conv_5_1 = nn.Conv2d(512, 512, 3, stride=1, padding=1)
        self.conv_5_2 = nn.Conv2d(512, 512, 3, stride=1, padding=1)
        self.conv_5_3 = nn.Conv2d(512, 512, 3, stride=1, padding=1)
        
        self.fc6 = nn.Linear(512 * 7 * 7, 4096)
        self.fc7 = nn.Linear(4096, 4096)
        self.fc8 = nn.Linear(4096, num_classes)

        
        
    
    # Load pretrained weight from .t7 file, which is not frequently used nowadays. 
    def load_weights(self, path="VGG_FACE.t7"):
        """ Function to load luatorch pretrained

        Args:
            path: path for the luatorch pretrained
        """
        model = torchfile.load(path, force_8bytes_long=True)
        counter = 1
        block = 1
        for i, layer in enumerate(model.modules):
            if layer.weight is not None:
                if block <= 5:
                    self_layer = getattr(self, "conv_%d_%d" % (block, counter))
                    counter += 1
                    if counter > self.block_size[block - 1]:
                        counter = 1
                        block += 1
                    self_layer.weight.data[...] = torch.tensor(layer.weight).view_as(self_layer.weight)[...]
                    self_layer.bias.data[...] = torch.tensor(layer.bias).view_as(self_layer.bias)[...]
                elif block < 8:
                    self_layer = getattr(self, "fc%d" % (block))
                    block += 1
                    self_layer.weight.data[...] = torch.tensor(layer.weight).view_as(self_layer.weight)[...]
                    self_layer.bias.data[...] = torch.tensor(layer.bias).view_as(self_layer.bias)[...]

                    
    # Forward step 
    def forward(self, x):
        """ Pytorch forward

        Args:
            x: input image (224x224)

        Returns: class logits

        """
        x = F.relu(self.conv_1_1(x))
        x = F.relu(self.conv_1_2(x))
        x1 = F.max_pool2d(x, 2, 2)
        
        
        
        x2 = F.relu(self.conv_2_1(x1))
        x2 = F.relu(self.conv_2_2(x2))
        x2 = F.max_pool2d(x2, 2, 2)
        
        x3 = F.relu(self.conv_3_1(x2))
        x3 = F.relu(self.conv_3_2(x3))
        x3 = F.relu(self.conv_3_3(x3))
        x3 = F.max_pool2d(x3, 2, 2)
        
        
        x4 = F.relu(self.conv_4_1(x3))
        x4 = F.relu(self.conv_4_2(x4))
        x4 = F.relu(self.conv_4_3(x4))
        x4 = F.max_pool2d(x4, 2, 2)
        
        
        x5 = F.relu(self.conv_5_1(x4))
        x5 = F.relu(self.conv_5_2(x5))
        x5 = F.relu(self.conv_5_3(x5))
        x5 = F.max_pool2d(x5, 2, 2)
        
        x6 = x5.view(x5.size(0), -1)
        x6 = F.relu(self.fc6(x6))
        x6 = F.dropout(x6, 0.5, self.training)
        
        x7 = F.relu(self.fc7(x6))
        x7 = F.dropout(x7, 0.5, self.training)
        
        return self.fc8(x7), x7, x6, x5, x4, x3, x2, x1
```

```python
# Load, split, and normalize your data
from os import listdir
from os.path import isfile, join
import numpy
import cv2
from sklearn.model_selection import train_test_split
from pprint import pprint
from torchvision import datasets, transforms, models
import skimage.transform as sktr
import matplotlib.pyplot as plt
 
# mypath='MMIeomtion CROP'
# onlyfiles = [ f for f in listdir(mypath) if isfile(join(mypath,f)) ]
# length = len(onlyfiles)
# images = numpy.empty(shape = (length,2), dtype= object)
# for n in range(0, len(onlyfiles)):
# #     print(cv2.imread(join(mypath,onlyfiles[n])).shape)
#     images[n,0] = cv2.resize(cv2.imread(join(mypath,onlyfiles[n])),(224,224)).astype(numpy.float32)
# #     images[n,0] = sktr.resize(cv2.imread(join(mypath,onlyfiles[n])),[224,224]).astype(numpy.float32)
#     images[n,0] = cv2.cvtColor(images[n,0], cv2.COLOR_BGR2RGB)
#     name = onlyfiles[n]
#     name = name.split('_')
# #     print(name)
#     if name[1] == 'Happiness':
#         images[n,1] = 0
#     else:
#         images[n,1] = 1
# #     print(name)

# X_train,X_test,y_train,y_test=train_test_split(
#         images[:,0],images[:,1],test_size=0.3,random_state=13)

# norm = numpy.array([129.1863, 104.7624, 93.5940], dtype = object)

# def normalise(data,norm):
#     for i in range(len(data)):
#         data[i] = data[i] - norm
#     return data
# # print(y_train)
# X_train_norm = normalise(X_train,norm)
# # X_train_norm = numpy.array(X_train_norm)
# X_test_norm = normalise(X_test,norm)
# # X_test_norm = numpy.array(X_test_norm)
# # print(X_train_norm.shape)


###############################################
import numpy as np
mypath = 'data'
def load_split_train_test(datadir, valid_size = .3):
#     transforms = transforms.Compose([transforms.Resize([244, 244]), transforms.ToTensor(), transforms.Normalize([129.1863, 104.7624, 93.5940], [255, 255, 255])])
    transform = transforms.Compose([transforms.Resize([224, 244]), transforms.ToTensor(), transforms.Normalize([129.1863/255, 104.7624/255, 93.5940/255], [1, 1, 1])])
    train_data = datasets.ImageFolder(datadir,       
                    transform=transform)
    test_data = datasets.ImageFolder(datadir,
                    transform=transform)
    num_train = len(train_data)
    indices = list(range(num_train))
    split = int(np.floor(valid_size * num_train))
    np.random.shuffle(indices)
    from torch.utils.data.sampler import SubsetRandomSampler
    train_idx, test_idx = indices[split:], indices[:split]
    train_sampler = SubsetRandomSampler(train_idx)
    test_sampler = SubsetRandomSampler(test_idx)
#     print(len(train_sampler))
#     print(len(test_sampler))
    trainloader = torch.utils.data.DataLoader(train_data,
                   sampler=train_sampler, batch_size=8)
    testloader = torch.utils.data.DataLoader(test_data,
                   sampler=test_sampler, batch_size=8)
    return train_data, trainloader, testloader
train_data, trainloader, testloader = load_split_train_test(mypath, .3)
# print(trainloader.dataset.classes)
# print(trainloader.dataset)
# print(len(trainloader))
# print(len(testloader))

##################################################
```
